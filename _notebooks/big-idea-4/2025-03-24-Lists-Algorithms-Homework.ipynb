# ‚úÖ Homework Hack #2: Python Code


```python
# Initialize a list with numbers from 1 to 30
numbers = list(range(1, 31))

# Filter numbers divisible by 3 but not by 5
filtered_numbers = [num for num in numbers if num % 3 == 0 and num % 5 != 0]

# Print the original and filtered lists
print("Original list:", numbers)
print("Filtered list (divisible by 3 but not by 5):", filtered_numbers)
```

```python
import pandas as pd

def filter_spotify_data(file_path):
    # Read the CSV file
    df = pd.read_csv(file_path)
    
    # Filter songs with over 10 million total streams
    filtered_df = df[df['Total Streams (Millions)'] > 10]
    
    # Display the filtered data clearly
    print("üéµ Songs with over 10 million streams:\n")
    print(filtered_df[['Artist', 'Album', 'Total Streams (Millions)']].to_string(index=False))
    
    return filtered_df

# Example usage
filtered_songs = filter_spotify_data("Spotify_2024_Global_Streaming_Data.csv")
```
